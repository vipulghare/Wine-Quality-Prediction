```python
plt.figure(figsize=(30,70))
m = 1
for i in A.columns:
    plt.subplot(A.shape[1],3,m)
    if(A[i].dtypes=="object"):
        sb.countplot(A[i])
    else:
        sb.distplot(A[i])
    m = m + 1
```

```python
#Declare Predictor and Target varibale
```

```python
Y=A[["quality"]]
```

```python
X = A.drop(labels=["quality"],axis = 1)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler #standarize the data
ss = StandardScaler()
X = pd.DataFrame(ss.fit_transform(X),columns=X.columns)
```

```python
X
```

```python
#split the data in training and testing data
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=20)
```

```python
# Fitting Random Forest Classifier to the Training set
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
model = rfc.fit(xtrain,ytrain)
pred_ts = model.predict(xtest)
pred_tr = model.predict(xtrain)
```

```python
#Calculation of Accuracy of the model
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy :",accuracy_score(ytrain,pred_tr))
print("Testing Accuracy :",accuracy_score(ytest,pred_ts))
```

```python
#Improve the accuracy of the model
```

```python
from sklearn.model_selection import GridSearchCV
tp = {"n_estimators":range(2,20,1),"max_depth":range(2,20,1),"criterion":["gini","entropy"]}

cv = GridSearchCV(rfc,tp,scoring="accuracy",cv=3)

cvmodel = cv.fit(xtrain,ytrain)
```

```python
#finding hyperparameters
```

```python
cvmodel.best_params_
```

```python
#again train the model
```

```python
rfc = RandomForestClassifier(criterion= 'entropy', max_depth=19, n_estimators= 16)
model = rfc.fit(xtrain,ytrain)
pred_ts = model.predict(xtest)
pred_tr = model.predict(xtrain)
```

```python
#Calculation of Accuracy of the model
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy :",accuracy_score(ytrain,pred_tr))
print("Testing Accuracy :",accuracy_score(ytest,pred_ts))
```
